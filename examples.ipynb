```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import pltvid

%config InlineBackend.figure_format = 'retina'
```

```python
camera = pltvid.Capture(dpi=290)

for x in np.arange(0,1,.1):
    fig, ax = plt.subplots(1,1,figsize=(3,2))
    ax.text(x, 0.5, "hi")
    camera.snap()

camera.save("/tmp/sample.gif", duration=200)
```
